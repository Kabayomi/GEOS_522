The importance of knowing the basic statistical values(mean, std, distribution, etc.) becomes apparent as you run multiple variations of the same analysis on the same data. It can greatly affect the outcome probabilities.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy import stats
from scipy.stats import iqr, kurtosis, skew, norm
```

```python
# Import depths1 into an array
df = pd.read_csv('depths1.txt', names=['Depth'])  # Import depth1.txt
depth1 = df.Depth.values                          # create an array of values from depths1.txt
df[0:5]                                           # check first 5 rows
```

```python
# Import depths2 into an array
df2 = pd.read_csv('depths2.txt', names=['Depth'])  # Import depth2.txt
depth2 = df2.Depth.values                          # create an array of values from depths2.txt
df2[:5]                                            # check first 5 rows
```

```python
# Combine both datasets into one, 2 column array
depthData = np.vstack((depth1,depth2))  # combine depth datasets into a single two column array
depthData[:5,:5]                        # check first 5 rows of both datasets
```

## 1. For each dataset, calculate and report the mean and standard deviation

### Depth1 Data

```python
d1Max = np.max(depth1)   # max
d1Min = np.min(depth1)   # min
d1men = np.mean(depth1)  # mean
d1men
```

```python
d1std = np.std(depth1)  # standard deviation
d1std
```

### Depth2 Data

```python
d2Max = np.max(depth2)   # max
d2Min = np.min(depth2)   # min
d2men = np.mean(depth2)  # mean
d2men
```

```python
d2std = np.std(depth2)  # standard deviation
d2std
```

## 2. For each dataset, calculate and report the median, mode, IQR, skewness, and kurtosis

### Depth1 Data

```python
d1med = np.median(depth1)   # median
d1mod = stats.mode(depth1)  # mode; [0] index is mode value
d1iqr = iqr(depth1)         # IQR
d1skw = skew(depth1)        # skew
d1kur = kurtosis(depth1)    # kurtosis

# Print stat variables
print('---Sample Stats for Depth1---')
print('Median: %.2f' % d1med)
print('Mode: %.2f' % d1mod[0])
print('IQR: %.2f' % d1iqr)
print('Skew: %.2f' % d1skw)
print('Kurtosis: %.2f' % d1kur)
```

### Depth2 Data

```python
d2med = np.median(depth2)   # median
d2mod = stats.mode(depth2)  # mode; [0] index is mode value
d2iqr = iqr(depth2)         # IQR
d2skw = skew(depth2)        # skew
d2kur = kurtosis(depth2)    # kurtosis

# Print stat variables
print('---Sample Stats for Depth2---')
print('Median: %.2f' % d2med)
print('Mode: %.2f' % d2mod[0])
print('IQR: %.2f' % d2iqr)
print('Skew: %.2f' % d2skw)
print('Kurtosis: %.2f' % d2kur)
```

```python
# QUESTION 9: WHAT IS THE PROBABILITY A VALUE WILL BE WITHIN 20CM OF THE AVERAGE VALUE?
# cdf(mean+20)-cdf(mean-20)


print('\033[1m'+'*Rough/Basic Probabilities; ignoring distributions*')
# Depth1 values
lenD1 = len(depth1)                     # depth1 length
d1_mu20 = depth1[depth1 <= (d1men-20)]  # depth1 values that are <=mean-20
d1_20mu = depth1[depth1 <= (d1men+20)]  # depth1 values that are <=mean+20
d1_9a = len(d1_mu20) / len(depth1)      # probability of depth1 values <=mean-20
d1_9b = len(d1_20mu) / len(depth1)      # probability of depth1 values <=mean+20
d1_9c = (d1_9b - d1_9a)*100             # probability of depth1 values +/-20cm of mean
print('\033[0m'+'Depth1 value is within +/-20cm of mean: %.2f' % d1_9c)
# Depth1 values
lenD2 = len(depth2)                     # depth2 length
d2_mu20 = depth2[depth2 <= (d2men-20)]  # depth2 values that are <=mean-20
d2_20mu = depth2[depth2 <= (d2men+20)]  # depth2 values that are <=mean+20
d2_9a = len(d2_mu20) / len(depth2)      # probability of depth2 values <=mean-20
d2_9b = len(d2_20mu) / len(depth2)      # probability of depth2 values <=mean+20
d2_9c = (d2_9b - d2_9a)*100             # probability of depth2 values +/-20cm of mean
print('\033[0m'+'Depth2 value is within +/-20cm of mean: %.2f' % d2_9c)


print('\033[1m'+'*Probabilities calculated through built-in CDF libraries*')
# Depth 1
d1_mu20_p1 = norm.cdf(d1men+20, loc=d1men, scale=d1std)  # probability of mu+20
d1_mu20_p2 = norm.cdf(d1men-20, loc=d1men, scale=d1std)  # probability of mu-20
d1_p9 = (d1_mu20_p1 - d1_mu20_p2)*100                    # probability of value between mu+20 and mu-20
print('\033[0m'+'Depth1 value being within +/-20cm of mean: %.2f percent' % d1_p9)
# Depth 2
d2_mu20_p1 = norm.cdf(d2men+20, loc=d2men, scale=d2std)  # probability of mu+20
d2_mu20_p2 = norm.cdf(d2men-20, loc=d2men, scale=d2std)  # probability of mu-20
d2_p9 = (d2_mu20_p1 - d2_mu20_p2)*100                    # probability of value between mu+20 and mu-20
print('\033[0m'+'Depth2 value being within +/-20cm of mean: %.2f percent' % d2_p9)
```

```python
# QUESTION 10: WHAT IS THE PROBABILITY A VALUE WILL BE AT LEAST 20CM ABOVE THE AVERAGE VALUE?
# cdf(max)-cdf(mean+20)

print('\033[1m'+'*Probabilities calculated through built-in libraries*')
# Depth 1
d1_max_p1 = norm.cdf(d1Max, loc=d1men, scale=d1std)      # probability of max value
d1_mu20_p2 = norm.cdf(d1men+20, loc=d1men, scale=d1std)  # probability of mu+20
d1_p10 = (d1_max_p1 - d1_mu20_p2)*100                    # probability of value between mu+20 and mu-20
print('\033[0m'+'Probability of a Depth1 value being at least 20cm above the mean: %.2f percent' % d1_p10)
# Depth 2
d2_max_p1 = norm.cdf(d2Max, loc=d2men, scale=d2std)      # probability of max value
d2_mu20_p2 = norm.cdf(d2men+20, loc=d2men, scale=d2std)  # probability of mu+20
d2_p10 = (d2_max_p1 - d2_mu20_p2)*100                    # probability of value between mu+20 and mu-20
print('\033[0m'+'Probability of a Depth2 value being at least 20cm above the mean: %.2f percent' % d2_p10)
```

```python
# QUESTION 11: WHAT IS THE PROBABILITY A VALUE WILL BE AT LEAST 20CM ABOVE THE AVERAGE VALUE?
# cdf(mean-20)-cdf(min)

print('\033[1m'+'*Probabilities calculated through built-in libraries*')
# Depth 1
d1_min_p1 = norm.cdf(d1Min, loc=d1men, scale=d1std)      # probability of min value
d1_mu20_p2 = norm.cdf(d1men-20, loc=d1men, scale=d1std)  # probability of mu+20
d1_p11 = (d1_mu20_p2 - d1_min_p1)*100                    # probability of value between mu-20 and min
print('\033[0m'+'Probability of a Depth1 value being at least 20cm below the mean: %.2f percent' % d1_p11)
# Depth 2
d2_min_p1 = norm.cdf(d2Min, loc=d2men, scale=d2std)      # probability of min value
d2_mu20_p2 = norm.cdf(d2men-20, loc=d2men, scale=d2std)  # probability of mu+20
d2_p11 = (d2_mu20_p2 - d2_min_p1)*100                    # probability of value between mu-20 and min
print('\033[0m'+'Probability of a Depth2 value being at least 20cm below the mean: %.2f percent' % d2_p11)
```
